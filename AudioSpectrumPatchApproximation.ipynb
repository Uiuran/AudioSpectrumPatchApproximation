###  Constant-Q Time-Frequency Analysis

Analyze 10s of a natural scene (field recording) using a log frequency scale (constant-Q transform) 
and plot the resulting time series and its time-averaged constant-Q spectrum.

```python
from pylab import *
rcParams['figure.figsize'] = (12.0, 8.0)
```

```python
from bregman.suite import *
F = LogFrequencySpectrum('chernobyl.wav',nhop=1024, nfft=8192, wfft=4096, npo=24) # constant-Q transform
subplot(211)
F.feature_plot(normalize=True, nofig=True, cmap=cm.hot) # plot the transform
title('Constant-Q Spectrum (Linear Amplitude)', fontsize=14)
subplot(212)
plot(log(1+F.X).sum(1)) # and the time-averaged constant-Q spectrum
title('Mean Constant-Q Spectrum: Log Amplitude ', fontsize=14)
xticks(arange(0,100,8), F._logfrqs[0:-1:8].round())
xlabel('Frequency (Hz)')
ylabel('Log Amplitude')
grid();axis('tight');ax=colorbar();ax.ax.set_visible(False)
```

### Inverting the analysis back to audio

Invert the constant-Q transform to an audio signal, using inverse constant-Q transform and inverse short-time Fourier transform.

```python
xh = F.inverse()
play(balance_signal(xh)) # Check the resynthesis without sparse coding
```

### Method 1. Sparse approximation dictionary learning, using mini bach gradient learning

Learn sparse codes from data using dictionary learnng on 16x16 patches of the constant-Q transform.

```python
import AudioSpectrumPatchApproximation as A # Casey's multi-scale 2D spectrogram patch sprase approximation library
reload(A) # for debugging
```

```python
SS = A.SparseApproxSpectrum(n_components=9, patch_size=(11,11)) # learn 9 components of 16x16 patches
SS.extract_codes(log(1+F.X), standardize=True) # Use standardized patches
SS.plot_codes(cbar=1, cmap=cm.hot) # show the learned codes
```

```python
SS.reconstruct_individual_spectra(plotting=1, cmap=cm.hot)
figure()
subplot(211)
feature_plot(F.X, normalize=True, nofig=True, cmap=cm.hot)
subplot(212)
feature_plot(exp(SS.X_hat)-1, normalize=True, nofig=True, cmap=cm.hot)
```

```python
X_hat = exp(SS.X_hat_l[0])-1 # <- change the patch to reconstruct here
x_hat = F.inverse(X_hat) #, Phi_hat=(rand(*F.STFT.shape)*2-1)*pi) 
play(balance_signal(x_hat))
feature_plot(X_hat,normalize=False, cmap=cm.hot)
```

<h2>Method 2: Decomposition with a "Gabor field" dictionary</h2>

```python
F = LogFrequencySpectrum('chernobyl.wav',nhop=1024, nfft=8192, wfft=4096, npo=24) # constant-Q transform
s1 = A.SparseApproxSpectrum(n_components=16, patch_size=(11,11))
s1.make_gabor_field(log(1+F.X), standardize=True)
s1.plot_codes(cbar=1,cmap=cm.hot) # show the learned codes
```

Reconstruct the constant-Q spectrogram using each learned patch basis. Do this for each patch separately.

```python
s1.reconstruct_individual_spectra(plotting=1, cmap=cm.hot)
figure()
subplot(211)
feature_plot(F.X, normalize=True, nofig=True, cmap=cm.hot)
subplot(212)
feature_plot(exp(s1.X_hat)-1, normalize=True, nofig=True, cmap=cm.hot)
```

```python
X_hat = exp(s1.X_hat_l[6])-1 # <- change the patch to reconstruct here
x_hat = F.inverse(X_hat) #, Phi_hat=(rand(*F.STFT.shape)*2-1)*pi) 
play(balance_signal(x_hat))
feature_plot(X_hat,normalize=True,cmap=cm.hot)
```
